```python
from sklearn import datasets
from sklearn.metrics import confusion_matrix
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from scipy.cluster.hierarchy import linkage, dendrogram

import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
iris
```

```python
kmeans = KMeans(n_clusters=3, n_init='auto')
kmeans.fit(iris.data)
print(kmeans.labels_)
```

```python
iris.target
```

```python
result = confusion_matrix(iris.target, kmeans.labels_)
print(result)
```

```python
dbscan = DBSCAN(eps=0.5, min_samples=3)
dbscan_labels = dbscan.fit_predict(iris.data)
print(dbscan_labels)
```

```python
agglo = AgglomerativeClustering(n_clusters=3)
agglo_labels = agglo.fit_predict(iris.data)
print(agglo_labels)
```

```python
result = confusion_matrix(iris.target, agglo_labels)
print(result)
```

```python
def plot_clusters(data, labels, title):
  colors = ['red', 'green', 'purple', 'black']
  plt.figure(figsize=(8,4))
  for i, c, l in zip(range(-1, 3), colors, ['Noise', 'Setosa', 'Versicolor', 'Virginica']):
    if i == -1:
      plt.scatter(data[labels == i, 0], data[labels == i, 3], c=colors[i], label=l, alpha=0.5, s=50, marker='x')
    else:
      plt.scatter(data[labels == i, 0], data[labels == i, 3], c=colors[i], label=l, alpha=0.5, s=50)
  plt.legend()
  plt.title(title)
  plt.xlabel('Sepal Length')
  plt.ylabel('Petal Width')
  plt.show()
```

```python
plot_clusters(iris.data, kmeans.labels_, 'KMeans')
```

```python
plot_clusters(iris.data, dbscan_labels, 'DBScan')
```

```python
plot_clusters(iris.data, agglo_labels, 'Agglomerative Clustering')
```

```python
plt.figure(figsize=(12, 6))
plt.title('Dendrogram')
plt.xlabel('Species')
plt.ylabel('Distance')
dendrogram(linkage(iris.data, method='ward'), truncate_mode='lastp', p=15)
plt.axhline(y=7, c='gray', lw=1, linestyle='dashed')
plt.show()
```
